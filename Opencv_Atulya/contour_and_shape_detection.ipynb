```python
import cv2
import numpy as np
```

```python
img = cv2.imread("assets//images//shapes.jpg")
cv2.imshow("original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray,30,150)
cont, hier = cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
```

```python
print(type(cont))
print(len(cont))
print(cont)
print(cont[0])
print(type(cont[0]))
print(cont[0].dtype)
```

```python
blank = np.zeros(img.shape,np.uint8)
cv2.drawContours(blank,cont,-1,(0,0,255),2)
cv2.imshow("blank",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blank = np.zeros(img.shape,np.uint8)
for cnt in cont:
    cv2.drawContours(blank,[cnt],-1,(255,0,0),2)
    cv2.imshow("blank",blank)
    if cv2.waitKey(10) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
blank = np.zeros(img.shape,np.uint8)
for cnt in cont:
    if cv2.contourArea(cnt) > 1000:
        cv2.drawContours(blank,[cnt],-1,(255,0,0),2)
cv2.imshow("blank",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
exam = cv2.imread("assets//images//rect.jpg")
cv2.imshow("example",exam)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
g = cv2.cvtColor(exam,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(g,150,255,cv2.THRESH_BINARY_INV)
cv2.imshow("examplethresh",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
exam2 = exam.copy()
c, h = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
c = max(c,key=cv2.contourArea)
cv2.drawContours(exam2,[c],-1,(255,0,0),2)
cv2.putText(exam2,f"no. of points {len(c)}",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
cv2.imshow("example",exam2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def any(x):
    pass
cv2.namedWindow("Contour Approximation")
cv2.resizeWindow("Contour Approximation",640,400)
cv2.createTrackbar("epsilon","Contour Approximation",0,100,any)
peri = cv2.arcLength(c,True)
while True:
    exam3 = exam.copy()
    eps = cv2.getTrackbarPos("epsilon","Contour Approximation")
    eps = eps/1000
    approx = cv2.approxPolyDP(c,peri*eps,True)
    cv2.drawContours(exam3,[approx],-1,(255,0,0),2)
    cv2.putText(exam3,f"no. of points {len(approx)}",(30,30),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    cv2.putText(exam3,f"epsilon {eps}",(60,60),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    cv2.imshow("Contour Approximation",exam3)
    if cv2.waitKey(10) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
l = 4
img2 = img.copy()
for cnt in cont:
    if cv2.contourArea(cnt) > 1000:
        peri = cv2.arcLength(cnt,True)
        approx = cv2.approxPolyDP(cnt,peri*0.04,True)
        if len(approx) == l:
            cv2.drawContours(img2,[approx],-1,(0,0,0),4)
            x,y,w,h = cv2.boundingRect(cnt)
            cv2.putText(img2,f" {len(cnt)}",(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2)
            cv2.rectangle(img2,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow("image",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
